```python
import ast
import re
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def clean_ingredients_column(value):
    """
    Converts ['a', 'b', 'c'] → 'a b c'
    and lightly lemmatizes each ingredient term.
    """
    # Step 1: If it's a stringified list, convert it to a real list
    if isinstance(value, str):
        try:
            value = ast.literal_eval(value)
        except (ValueError, SyntaxError):
            # In case it's already just a string
            value = [value]
    
    # Step 2: Join into one string
    if isinstance(value, list):
        text = ' '.join(map(str, value))
    else:
        text = str(value)
    
    # Step 3: Light clean — keep letters only and lowercase
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Step 4: Light lemmatization (one pass)
    text = ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
    
    return text
```

```python
df['ingredients'] = df['ingredients'].apply(clean_ingredients_column)
```

```python
import pandas as pd
# Convert list-like strings (if stored as text) to real Python lists
import ast
def safely_eval_list(x):
    if isinstance(x, str) and x.startswith('[') and x.endswith(']'):
        try:
            return ast.literal_eval(x)
        except:
            return [x]
    elif isinstance(x, list):
        return x
    else:
        return [str(x)]

# Apply to directions column
df['directions'] = df['directions'].apply(safely_eval_list)

# Join list elements into one string separated by space
df['directions'] = df['directions'].apply(lambda x: ' '.join(x))
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains import RetrievalQA

# initialize Gemini
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-pro",
    google_api_key=GOOGLE_API_KEY,
    temperature=0.3
)
```

```python
rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    return_source_documents=True
)
```

```python
query = "I have eggs and tomatoes, what can I cook?"
response = rag_chain.invoke({"query": query})

print("🧠 Question:", query)
print("\n🍳 Answer:", response['result'])
```
