```python
db_name = "Cooking_Assistant"
collection_name = "recepie_data"
collection_url = "mongodb+srv://thoufeq:thoufeq@cluster0.sj2zy7w.mongodb.net/?appName=Cluster0"
```

```python
import pymongo
```

```python
from pymongo.errors import BulkWriteError
```

```python
batch_size = 1000
```

```python
client = pymongo.MongoClient(collection_url)
data_base = client[db_name]
collection = data_base[collection_name]

for i in range(0, len(data), batch_size):
    try:
        collection.insert_many(data[i:i + batch_size], ordered=False)
    except BulkWriteError as e:
        print("Batch insert error:", e.details)
```
